```python
# slide 7
n = 100
lambda_ = 2
U = np.random.uniform(0,1,n)
X = -np.log(1-U) / lambda_
print(X)

plt.figure(figsize=(6,4))
plt.hist(X, bins=15, color="skyblue", edgecolor="white", alpha=0.85)
plt.title("Histogram of Exponential Samples (λ = 2)", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

### Part 3

```python
# slide 11
counter = 0
num_sims = 2000
U = []
A = []

for i in range(num_sims):
    u = np.random.uniform()
    d = np.random.uniform()
    U.append(u)
    if d < u:
        A.append(u)
        counter += 1

print(counter)

plt.figure(figsize=(6,4))
plt.hist(U, bins=15, color="skyblue", edgecolor="white", alpha=0.85)
plt.title("Uniform Distribution - Histogram of U", fontsize=14)
plt.xlabel("U", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
plt.hist(A, bins=15, color="skyblue", edgecolor="white", alpha=0.85)
plt.title("Conditional Distribution - Histogram of A", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# slide 17
n = 100
c = 3.8
target_samples = []
counter = 0

# Accept-reject sampling
while counter < n:
    U = np.random.uniform(size=2)
    proposal = np.tan(np.pi * U[0])
    f = 1 / (1 + proposal**4)
    g = 1 / (np.pi * (1 + proposal**2))
    
    if U[1] < (f / (c * g)):
        target_samples.append(proposal)
        counter += 1

plt.figure(figsize=(6, 4))
plt.hist(target_samples, bins=20, color="skyblue", edgecolor="white", alpha=0.85)
plt.title("Accepted Samples from Target Distribution", fontsize=14)
plt.xlabel("Proposal Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# slide 2
n = 1000
counter = 0
target_sample = []
mu, sigma = 0, 1

while counter < n:
    proposal = np.random.normal(loc=mu, scale=sigma)
    if proposal > 0:
        target_sample.append(proposal)
        counter += 1

plt.figure(figsize=(6, 4))
plt.hist(target_sample, bins=30, color="skyblue", edgecolor="white", alpha=0.85)
plt.title("Conditional Distribution, X~N(0,1)|X>=0", fontsize=14)
plt.xlabel("Proposal Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# slide 3
n = 1000
mu, sigma = 0, 1
samples = np.random.normal(loc=mu, scale=sigma, size=n*2)
target_sample = samples[samples > 0]
target_sample = target_sample[:n]

plt.figure(figsize=(6, 4))
plt.hist(target_sample, bins=30, color="skyblue", edgecolor="white", alpha=0.85)
plt.title("Conditional Distribution, X~N(0,1)|X>=0", fontsize=14)
plt.xlabel("Proposal Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# slide 4
n = 1000
X = np.random.normal(size=n)
Z = norm.pdf(X)
Y = np.random.uniform(0, Z)

plt.figure(figsize=(7, 5))
plt.scatter(X, Y, color='skyblue', s=20, alpha=0.7, edgecolor='k')  # filled dots with alpha
plt.title('Uniform Samples Under Standard Normal PDF', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Part 6

```python
# slide 5
n = 1000
m = n // 2
X = np.zeros(n)
R = np.zeros(m)
U = np.random.uniform(0,1,m)
Theta = np.random.uniform(0, 2*np.pi, m)

# Box-Muller transform
for i in range(m):
    R[i] = np.sqrt(-2 * np.log(U[i]))
    X[2*i] = R[i] * np.cos(Theta[i])
    X[2*i+1] = R[i] * np.sin(Theta[i])

# Plot histogram nicely
plt.figure(figsize=(7,5))
plt.hist(X, bins=30, color='lightcoral', edgecolor='white', alpha=0.85)
plt.title('Histogram of Standard Normal Samples (Box-Muller)', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
